```python
import pandas as pd
```

```python
df = pd.read_csv('netflix_titles.csv')
df.head(2)
```

```python
df = df[df['type'] == 'Movie']
```

```python
import re
def get_first_nums(s):
    return re.findall(r'\d+', s)[0]

df['duration'] = pd.to_numeric(df['duration'].apply(get_first_nums))
```

```python
df.duration.plot(kind = 'hist', title = 'Продолжительность фильмов, мин.')
```

Распределение продолжительности фильмов является нормальным. 

```python
ax = df.groupby('release_year').mean()['duration'].plot(title = 'Средняя продолжительность фильмов по годам')
```

Средняя продолжительность фильмов достигла своего пика в середине 60х годов.

```python
df.plot(kind = 'scatter', x='release_year', y='duration')
```

Из графика видно что после середины 60х наблюдалось увеличение колтчества фильмов, что оказало сильное влияние на показатель средней продолжительности. Сейчас средняя продолжительность имеет не высокое значение, но разнообразие продолжительностей гораздо шире. 

```python
df.groupby('rating').count()['show_id'].plot(kind = 'bar', title = 'Количество фильмов по рейтингам')
```

Наибольшее кол-во фильмов имеет рейтинг TV-MA.
