```python
horse_data = horse_data.replace('?',np.NaN)
horse_data['Age'] = horse_data['Age'].astype('object')
horse_data['Hospital Number'] = horse_data['Hospital Number'].astype('object')
horse_data['rectal temperature'] = pd.to_numeric(horse_data['rectal temperature'])
horse_data['pulse'] = pd.to_numeric(horse_data['pulse'])
horse_data['total protein'] = pd.to_numeric(horse_data['total protein'])
horse_data['respiratory rate'] = pd.to_numeric(horse_data['respiratory rate'])

horse_data.info()
```

```python
horse_data.describe()
```

```python
def remove_outlies(df, row):
    q1 = df[row].quantile(0.25)
    q3 = df[row].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - (1.5*iqr)
    upper_bound = q3 + (1.5*iqr)
    remove_outliers_df = df[(df[row].between(lower_bound,upper_bound, inclusive = True))|(pd.isnull(df[row]))]
    removed = pd.concat([df, remove_outliers_df]).drop_duplicates(keep = False)
    return remove_outliers_df, removed
```

```python
_,removed = remove_outlies(horse_data, 'rectal temperature')
removed.describe()
```
