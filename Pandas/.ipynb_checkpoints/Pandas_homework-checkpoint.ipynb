Задание 2

По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
data3 = pd.read_csv('power.csv')
data3.head()
```

```python
data3 = data3[(data3.country == 'Latvia') | (data3.country == 'Lithuania') | (data3.country == 'Estonia')]
data3.head()
```

```python
data3 = data3[(data3.category == 4) | (data3.category == 12) | (data3.category == 21)]
data3.head()
```

```python
data3 = data3[(2005<=data3.year) & (data3.year<=2010)]
data3.head()
```

```python
data3 = data3[ data3.quantity >=0 ]
data3.quantity.sum()
```
