# Домашнее задание к лекции "Функции"

Вам нужно помочь секретарю автоматизировать работу. Для этого нужно написать программу, которая будет на основе хранимых данных исполнять пользовательские команды.

Исходные данные имеют следующую структуру:

1. перечень всех документов
```
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
```
2. перечень полок, на которых хранятся документы (если документ есть в documents, то он обязательно должен быть и в directories)
```
directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}
```

Общие требования к программе:
- код должен быть грамотно декомпозирован (каждая функция отвечает за свою конкретную задачу, дублирующийся функционал переиспользуется, а его код не повторяется);
- в коде отсутствуют глобальные переменные (за исключением **documents** и **directories**);
- пользовательский ввод обрабатывается в цикле **while** до тех пор, пока пользователь явно не завершит программу (вводом команды "**q**").

````python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}

def p():
    doc_number=input('Введите номер документа:')
    for document in documents:
        if document['number'] == doc_number:
            print('Владелец документа: '+ document['name'])
            
def s():
    doc_number=input('Введите номер документа:')
    i = False
    for shelf, doc_list  in directories.items():
        if doc_number in doc_list:
            print(f'Документ хранится на полке: {shelf}')
            i = True
    if i == False:
        print('Документ не найден в базе')
    
def l():
    print('Текущий список документов:')
    for document in documents:
        for shelf, doc_list  in directories.items():
            if document['number'] in doc_list:
                print('№: '+ document['number']+', тип: '+ document['type']+', владелец: ' + document['name'] + ', полка хранения: '+ shelf)

def ads():
    shelf=input('Введите номер полки:')
    if shelf in directories.keys():
        string = 'Такая полка уже существует. Текущий перечень полок: '
        for key in directories.keys():
            string+=key+', '
        print(string[0:-2]+'.')
    else: 
        directories[shelf]=[]
        string = 'Полка добавлена. Текущий перечень полок: '
        for key in directories.keys():
             string+=key+', '
        print(string[0:-2]+'.')
        
def ds():
    shelf=input('Введите номер полки:')
    if shelf in directories.keys():
        if len(directories[shelf]) == 0:
            del directories[shelf]
            string='Полка удалена. Текущий перечень полок: '
            for key in directories.keys():
                string+=key+', '
            print(string[0:-2]+'.')
        else: 
            string='На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: '
            for key in directories.keys():
                string+=key+', '
            print(string[0:-2]+'.')
    else:
        string='Такой полки не существует. Текущий перечень полок: '
        for key in directories.keys():
            string+=key+', '
        print(string[0:-2]+'.')
````

````python
def ad():
    doc_num=input('Введите номер документа: ')
    doc_type=input('Введите тип документа: ')
    name=input('Введите владельца документа: ')
    shelf=input('Введите полку для хранения: ')
    if shelf in directories:
        documents.append({'type': doc_type, 'number': doc_num, 'name': name})
        directories[shelf].append(doc_num)
    else:
        print('Такой полки не существует. Добавьте полку командой as.')
    l()

def d():
    doc_num=input('Введите номер документа: ')
    del_num = None
    for i, doc_dict in enumerate(documents):
        if doc_dict['number'] == doc_num:
            del_num=i
            break
    if del_num != None:
        documents.pop(del_num)
        for value in directories.values():
            if doc_num in value:
                value.remove(doc_num)
        print('Документ удален.')
    else:
        print('Документ не найден в базе.')
    l()

def m():
    doc_num=input('Введите номер документа: ')
    shelf=input('Введите номер полки: ')
    
    if shelf not in directories:
        print('Такой полки не существует.')
        
    else:
        i = False
        for value in directories.values():
            if doc_num in value:
                value.remove(doc_num)
                i=True
                directories[shelf].append(doc_num)
        if i==False: print('Документ не найден в базе.')
    l()
````

````python
while True:
    command = input('Введите команду:')
    if command == 'p': p()
    if command == 's': s()
    if command == 'l': l()
    if command == 'ads': ads()
    if command == 'ds': ds()
    if command == 'ad': ad()
    if command == 'd': d()
    if command == 'm': m()
    elif command == 'q': break
````
