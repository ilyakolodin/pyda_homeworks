### 7. Проверить гипотезу, что сотрудники с высоким окладом проводят на работе больше времени, чем сотрудники с низким окладом.

```python
data_low = np.array([item[0] for item in data[data['salary'] == 'low'][['average_montly_hours']].values])
data_high = np.array([item[0] for item in data[data['salary'] == 'high'][['average_montly_hours']].values])
stats.ttest_ind(data_low, data_high)
```

Значение p-value велико, что не позволяет отклонить нулевую гипотезу равенства средних значений выборок. Нельзя сказать, что люди с большей ЗП проводят больше времени на работе.

### 9. Разделить данные на тестовую и обучающую выборки 
### Построить модель LDA, предсказывающую уволился ли сотрудник на основе имеющихся факторов (кроме department и salary)
### Оценить качество модели на тестовой выборке

```python
data_class = np.array(data['left'])
data = data.drop(['department', 'salary', 'left'], axis='columns')
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, data_class, test_size=0.25)
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
lda = LinearDiscriminantAnalysis()
lda.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, lda.predict(X_test))
```
